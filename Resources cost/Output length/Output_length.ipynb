```python
import torch
import matplotlib.pyplot as plt
import numpy as np
```

```python
y_length = np.arange(2, 400, 20)
y_length_nsde = np.arange(2, 400, 20)
y_length_nsde[0] = 3
```

```python
y_length_memory_nsde_1 = torch.load('nsde_sigwgan/y_length_memory.pt')
y_length_time_nsde_1 = torch.load('nsde_sigwgan/y_length_time.pt')

y_length_memory_nsde_2 = torch.load('nsde_sigwgan/y_length_memory_2.pt')
y_length_time_nsde_2 = torch.load('nsde_sigwgan/y_length_time_2.pt')

y_length_memory_lstmsig = torch.load('lstm_sigwgan/y_length_memory.pt')
y_length_time_lstmsig = torch.load('lstm_sigwgan/y_length_time.pt')

y_length_memory_lstm = torch.load('lstm_wgan/y_length_memory.pt')
y_length_time_lstm = torch.load('lstm_wgan/y_length_time.pt')
```

```python
plt.rcParams['figure.figsize'] = [14, 5]
fig, axs = plt.subplots(1,2)

plt.rcParams['figure.figsize'] = [8, 5]
a = axs[0].plot(y_length, y_length_memory_lstm, label='LSTM CWGAN', linewidth=2.5, alpha=1, ls='dotted', color='green')
a = axs[0].plot(y_length[:16], y_length_memory_lstmsig[:16], label='LSTM CSig-WGAN', linewidth=2.5, alpha=1, ls='dotted', color='red')
a = axs[0].plot(y_length_nsde, y_length_memory_nsde_1, label='NSDE CSig-WGAN', linewidth=2.5, alpha=1, ls='--', color='blue')
#a = axs[0].plot(y_length_nsde, y_length_memory_nsde_2, label='NSDE CSig-WGAN 2', linewidth=2.5, alpha=1, ls='--', color='orange')

a = axs[0].legend(fontsize=13)
a = axs[0].grid()
a = axs[0].set_xlabel('Length of the output stream', fontsize=13)
a = axs[0].set_ylabel('Memory (MB)', fontsize=13)

a = axs[1].plot(y_length, y_length_time_lstm, label='LSTM CWGAN', linewidth=2.5, alpha=1, ls='dotted', color='green')
a = axs[1].plot(y_length[:16], y_length_time_lstmsig[:16], label='LSTM CSig-WGAN', linewidth=2.5, zorder=4, ls='dotted', color='red')
a = axs[1].plot(y_length_nsde, y_length_time_nsde_1, label='NSDE CSig-WGAN', linewidth=2.5, zorder=2, ls='--', color='blue')
#a = axs[1].plot(y_length_nsde, y_length_time_nsde_2, label='NSDE CSig-WGAN 2', linewidth=2.5, ls='--', color='orange')


a = axs[1].legend(fontsize=13)
a = axs[1].grid()
a = axs[1].set_xlabel('Length of the output stream', fontsize=13)
a = axs[1].set_ylabel('Time per step (s)', fontsize=13)
```
