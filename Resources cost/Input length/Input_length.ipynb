```python
import torch
import matplotlib.pyplot as plt
import numpy as np
```

```python
x_length = np.arange(2, 400, 20)
```

```python
x_length_memory_nsde_1 = torch.load('nsde_sigwgan/x_length_memory.pt')
x_length_time_nsde_1 = torch.load('nsde_sigwgan/x_length_time.pt')

x_length_memory_nsde_2 = torch.load('nsde_sigwgan/x_length_memory_2.pt')
x_length_time_nsde_2 = torch.load('nsde_sigwgan/x_length_time_2.pt')

x_length_memory_lstmsig = torch.load('lstm_sigwgan/x_length_memory.pt')
x_length_time_lstmsig = torch.load('lstm_sigwgan/x_length_time.pt')

x_length_memory_lstm = torch.load('lstm_wgan/x_length_memory.pt')
x_length_time_lstm = torch.load('lstm_wgan/x_length_time.pt')
```

```python
plt.rcParams['figure.figsize'] = [14, 5]
fig, axs = plt.subplots(1,2)

plt.rcParams['figure.figsize'] = [8, 5]
a = axs[0].plot(x_length, x_length_memory_lstm, label='LSTM CWGAN', linewidth=2.5, alpha=1, ls='dotted', color='green')
a = axs[0].plot(x_length[:18], x_length_memory_lstmsig, label='LSTM CSig-WGAN', linewidth=2.5, alpha=1, ls='dotted', color='red')
a = axs[0].plot(x_length, x_length_memory_nsde_1, label='NSDE CSig-WGAN 1', linewidth=2.5, alpha=1, ls='--', color='blue')
a = axs[0].plot(x_length, x_length_memory_nsde_2, label='NSDE CSig-WGAN 2', linewidth=2.5, alpha=1, ls='--', color='orange')

a = axs[0].legend(fontsize=13)
a = axs[0].grid()
a = axs[0].set_xlabel('Length of the input stream', fontsize=13)
a = axs[0].set_ylabel('Memory (MB)', fontsize=13)

a = axs[1].plot(x_length, x_length_time_lstm, label='LSTM CWGAN', linewidth=2.5, alpha=1, ls='dotted', color='green')
a = axs[1].plot(x_length[:18], x_length_time_lstmsig[:18], label='LSTM CSig-WGAN', linewidth=2.5, zorder=4, ls='dotted', color='red')
a = axs[1].plot(x_length, x_length_time_nsde_1, label='NSDE CSig-WGAN 1', linewidth=2.5, zorder=2, ls='--', color='blue')
a = axs[1].plot(x_length, x_length_time_nsde_2, label='NSDE CSig-WGAN 2', linewidth=2.5, ls='--', color='orange')


a = axs[1].legend(fontsize=13)
a = axs[1].grid()
a = axs[1].set_xlabel('Length of the input stream', fontsize=13)
a = axs[1].set_ylabel('Time per step (s)', fontsize=13)
```

```python
a = plt.plot(x_length[:18], x_length_time_lstmsig[:18], label='LSTM Sig-WGAN', linewidth=2.5, zorder=4, ls='--', color='red')
a = plt.plot(x_length, x_length_time_nsde_1, label='NSDE Sig-WGAN 1', linewidth=2.5, zorder=2, ls='--', color='blue')
a = plt.plot(x_length, x_length_time_nsde_2, label='NSDE Sig-WGAN 2', linewidth=2.5, ls='--', color='orange')


a = plt.legend(fontsize=13)
a = plt.grid()
a = plt.xlabel('Length of the input stream', fontsize=13)
a = plt.ylabel('Time per step (s)', fontsize=13)
```
